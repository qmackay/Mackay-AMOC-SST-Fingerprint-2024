```python
import netCDF4
import numpy as np
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs
from matplotlib import pyplot as plt
import scipy.stats as sp
```

```python
## File_Dir's

histSST = xr.open_dataarray('[HIST WH INDEX PATH]')

histAMOCData = xr.open_dataarray('[HIST AMOC INDEX PATH]')

# Figure Names and Titles

fig_title = ['None'] * 10

fig_title[0] = 'WH INDEX SSP vs AMOC INDEX - RLZN 1'
fig_title[1] = 'WH INDEX SSP vs AMOC INDEX - RLZN 2'
fig_title[2] = 'WH INDEX SSP vs AMOC INDEX - RLZN 3'
fig_title[3] = 'WH INDEX SSP vs AMOC INDEX - RLZN 4'
fig_title[4] = 'WH INDEX SSP vs AMOC INDEX - RLZN 5'
fig_title[5] = 'WH INDEX SSP vs AMOC INDEX - RLZN 6'
fig_title[6] = 'WH INDEX SSP vs AMOC INDEX - RLZN 7'
fig_title[7] = 'WH INDEX SSP vs AMOC INDEX - RLZN 8'
fig_title[8] = 'WH INDEX SSP vs AMOC INDEX - RLZN 9'
fig_title[9] = 'WH INDEX SSP vs AMOC INDEX - RLZN 10'

fig_name = ['None'] * 10

fig_name[0] = 'WH Scatter Plot RLZN 1'
fig_name[1] = 'WH Scatter Plot RLZN 2'
fig_name[2] = 'WH Scatter Plot RLZN 3'
fig_name[3] = 'WH Scatter Plot RLZN 4'
fig_name[4] = 'WH Scatter Plot RLZN 5'
fig_name[5] = 'WH Scatter Plot RLZN 6'
fig_name[6] = 'WH Scatter Plot RLZN 7'
fig_name[7] = 'WH Scatter Plot RLZN 8'
fig_name[8] = 'WH Scatter Plot RLZN 9'
fig_name[9] = 'WH Scatter Plot RLZN 10'

# R**2 Data Storage

r_squared_storage = ['None'] * 10
a_value_storage = ['None'] * 10
b_value_storage = ['None'] * 10
x_storage = ['None'] * 10
y_storage = ['None'] * 10
```

```python
for i in range(0,10):

    #AMOC Calculations

    specific_msftmz_rlzn = histAMOCData[i]
    divide_amoc = specific_msftmz_rlzn/1e10
    amoc_anomaly = divide_amoc

    # Specific Rlzn SST
    hist_rlzn = histSST[i]

    # Data Assignment and Plot Generation

    y = np.array(amoc_anomaly)
    x = np.array(histSST[i])

    x_storage[i] = x
    y_storage[i] = y
 
    #linear regression and variables

    a,b,r,p,err= sp.linregress(hist_rlzn, amoc_anomaly)
    print("The M value is:", a)
    print("The B value is:", b)
    print("The R value is:", r**2)
    print("The P value is:", p)
    print("The stderror is", err)
    r_squared_storage[i] = r
    a_value_storage[i] = a
    b_value_storage[i] = b

    plt.scatter(x, y, color = "k",
                marker = "o", s = 2)
  
    # plotting the regression line
    plt.plot(x, a*x+b, color = "b")
    
    # putting labels
    plt.xlabel('WH Index Anomaly', fontsize = 13)
    plt.ylabel('AMOC Index Anomaly (1e10)', fontsize = 13)
    plt.rcParams.update({'font.family':'serif'})
    plt.rcParams.update({'font.sans-serif':'Times'})
    plt.rcParams.update({'font.size': 13})
    plt.minorticks_on()

    # function to show plot
    r_plot = "R^2 = ", '%.3f'%(r**2)
    plt.title(fig_title[i], y=1.05)
    fig_dir = '[EXPORT DIRECTORY]'
    fig_namesave = fig_name[i]
    plt.savefig(fig_dir + fig_namesave, dpi = 600, facecolor='white', transparent=False)
    plt.show()
```
