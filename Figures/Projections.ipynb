```python
# running just 1 realization to standardize

# calculating DIFFERENCE between PROJECTED AND ACTUAL

import scipy.stats as sp

# ######################################## SSP ########################################

all_ssp_rlzn_diff =['none']*1

# Loading and calculating the difference between the projected SST and the AMOC value

for i in range(0,1):
    ssp_amoc = amoc_ssp_index.msftmz/1e9
    ssp_amoc_rlzn = ssp_amoc[i]
    ssp_amoc_rlzn = ssp_amoc_rlzn.values

    ssp_sst = ssp_sst_projected_plot[i]
    ssp_sst_rlzn = ssp_sst.load()
    ssp_sst_rlzn = ssp_sst_rlzn.values

    ssp_rlzn_diff_measurement = ['none'] * 85

    for k in range(0,85):
        ssp_rlzn_diff_measurement[k] = ssp_sst_rlzn[k] - ssp_amoc_rlzn[k]

    all_ssp_rlzn_diff[i] = ssp_rlzn_diff_measurement

# Loading and calculating the difference between the projected SST and the AMOC value

# ######################################## Historical ########################################

all_hist_rlzn_diff = ['none'] * 1

# Loading and calculating the difference between the projected SST and the AMOC value

for i in range(0,1):
    rlzn_amoc = amoc_hist_index.msftmz/1e9
    rlzn_amoc = rlzn_amoc[i]
    rlzn_amoc = rlzn_amoc.values

    rlzn_sst = hist_sst_projected_plot[i]
    rlzn_sst = rlzn_sst.load()
    rlzn_sst = rlzn_sst.values

    rlzn_diff_measurement = ['none'] * 165

    for k in range(0,165):
        rlzn_diff_measurement[k] = rlzn_sst[k] - rlzn_amoc[k]

    all_hist_rlzn_diff[i] = rlzn_diff_measurement

# ######################################## Abrupt ########################################

abrupt_sst_projected = abrupt_ssta3.load()

abrupt_diff_measurement = ['none'] * 165

# running full

for j in range(0,165):
    abrupt_diff_measurement[j] = abrupt_sst_projected.values[:,j] - abrupt_amoc_index.values[j]

# running just first 50 years

abrupt_diff_measurement_first50 = ['none'] * 50

for j in range(0,50):
    abrupt_diff_measurement_first50[j] = abrupt_sst_projected.values[:,j] - abrupt_amoc_index.values[j]

# running regime 2

abrupt_diff_measurement_regime2 = ['none'] * 115

for j in range(50,165):
    abrupt_diff_measurement_regime2[j-50] = abrupt_sst_projected.values[:,j] - abrupt_amoc_index.values[j]
```

```python
#running just 1 realization

# calculate real RMSD

#define RMSD calculation
def rmsd(supplied_diff_measurement):
    sample_length = len(supplied_diff_measurement)
    supplied_diff_measurement_squared = np.zeros(sample_length)
    for i in range(0,sample_length):
        supplied_diff_measurement_squared[i] = supplied_diff_measurement[i]**2

    rmsd_value = np.sqrt((sum(supplied_diff_measurement_squared))/sample_length)

    return rmsd_value


rmsd_abrupt_full = rmsd(abrupt_diff_measurement)
print("This is the RMSD for Abrupt Full:", rmsd_abrupt_full)

rmsd_abrupt_first50 = rmsd(abrupt_diff_measurement_first50)
print("This is the RMSD for Abrupt First50:", rmsd_abrupt_first50)

rmsd_abrupt_regime2 = rmsd(abrupt_diff_measurement_regime2)
print("This is the RMSD for Abrupt Regime 2:", rmsd_abrupt_regime2)

hist_rmsd = np.zeros(1)
for i in range(0,1):
    hist_rmsd[i] = rmsd(all_hist_rlzn_diff[i])

total_hist_rmsd = sum(hist_rmsd) / 1
print("This is the RMSD for Full Historical:", total_hist_rmsd)

ssp_rmsd = np.zeros(1)
for i in range(0,1):
    ssp_rmsd[i] = rmsd(all_ssp_rlzn_diff[i])

total_ssp_rmsd = sum(ssp_rmsd) / 1
print("This is the RMSD for SSP245:", total_ssp_rmsd)

####################################
print("  ")
print("  ")

#calculate as percentage compared to standard deviation of ONLY OBSERVED

rmsd_ratio_abruptfull = rmsd_abrupt_full / sp.tstd(abrupt_amoc_index.values)
print("The RMSD / StDev(Observed) value for Full Abrupt is:", rmsd_ratio_abruptfull)

rmsd_ratio_abruptfirst50 = rmsd_abrupt_first50 / sp.tstd(abrupt_amoc_index.values[0:50])
print("The RMSD / StDev(Observed) value for Abrupt First 50 is:", rmsd_ratio_abruptfirst50)

rmsd_ratio_abruptregime2 = rmsd_abrupt_regime2 / sp.tstd(abrupt_amoc_index.values[50:])
print("The RMSD / StDev(Observed) value for Abrupt First 50 is:", rmsd_ratio_abruptfirst50)

#hist
rmsd_ratio_historical_rlzns = np.zeros(1)
for i in range(0,1):
    rlzn_amoc = amoc_hist_index.msftmz/1e9
    rmsd_ratio_historical_rlzns[i] = hist_rmsd[i] / sp.tstd(rlzn_amoc[i,:])

rmsd_ratio_historical = sum(rmsd_ratio_historical_rlzns) / 1
print("The RMSD / StDev(Observed) value for Historical is:", rmsd_ratio_historical)

#SSP245
rmsd_ratio_ssp_rlzns = np.zeros(1)
for i in range(0,1):
    ssp_amoc = amoc_ssp_index.msftmz/1e9
    rmsd_ratio_ssp_rlzns[i] = ssp_rmsd[i] / sp.tstd(ssp_amoc[i,:])

rmsd_ratio_ssp = sum(rmsd_ratio_ssp_rlzns) / 1
print("The RMSD / StDev(Observed) value for SSP245 is:", rmsd_ratio_ssp)


####################################
print(" ")
print(" ")

#calculate as a fraction of the mean value

# abrupt full
avg_amoc_abruptfull = (sum(abrupt_amoc_index.values) / 165)
print("Average AMOC for Abrupt Full", avg_amoc_abruptfull)
rmsd_mean_comp_abruptfull = rmsd_abrupt_full / avg_amoc_abruptfull
print("RMSD / average AMOC Value for Abrupt Full =", rmsd_mean_comp_abruptfull)


# abrupt first 50
avg_amoc_abruptfirst50 = (sum(abrupt_amoc_index.values[0:50]) / 50)
print("Average AMOC for Abrupt First 50", avg_amoc_abruptfirst50)
rmsd_mean_comp_abruptfirst50 = rmsd_abrupt_first50 / avg_amoc_abruptfirst50
print("RMSD / average AMOC Value for Abrupt First 50 =", rmsd_mean_comp_abruptfirst50)

# abrupt Regime 2
avg_amoc_abruptregime2 = (sum(abrupt_amoc_index.values[50:]) / 115)
print("Average AMOC for Abrupt Regime 2", avg_amoc_abruptregime2)
rmsd_mean_comp_abruptregime2 = rmsd_abrupt_regime2 / avg_amoc_abruptregime2
print("RMSD / average AMOC Value for Abrupt Regime 2 =", rmsd_mean_comp_abruptregime2)


# historical
rmsd_mean_comp_historical = np.zeros(1)
hist_avg_amoc = np.zeros(1)
for i in range(0,1):
    rlzn_amoc = amoc_hist_index.msftmz/1e9
    rlzn_amoc = (sum(rlzn_amoc[i,:]) / 165)
    rmsd_mean_comp_historical[i] = (hist_rmsd[i] / rlzn_amoc)
    hist_avg_amoc[i] = rlzn_amoc
rmsd_mean_comp_historical_allrlzn = sum(rmsd_mean_comp_historical) / 1
print("Average AMOC for Historical", np.average(hist_avg_amoc, axis=0))
print("RMSD / average AMOC Value for Historical =", rmsd_mean_comp_historical_allrlzn)

#SSP245
rmsd_mean_comp_ssp = np.zeros(1)
ssp_avg_amoc = np.zeros(1)
for i in range(0,1):
    ssp_amoc = amoc_ssp_index.msftmz/1e9
    ssp_amoc = (sum(ssp_amoc[i,:]) / 85)
    rmsd_mean_comp_ssp[i] = (ssp_rmsd[i] / ssp_amoc)
    ssp_avg_amoc[i] = ssp_amoc
rmsd_mean_comp_ssp_allrlzn = sum(rmsd_mean_comp_ssp) / 1
print("Average AMOC for SSP", np.average(ssp_avg_amoc, axis=0))
print("RMSD / average AMOC Value for SSP245 =", rmsd_mean_comp_ssp_allrlzn)
```
